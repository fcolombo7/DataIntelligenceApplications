#### Comparison with the task_3

```python
task3 = Task3(data_generator)
task3.load('simulations_results/result_Step#3_2.zip')
```

```python
context_result = task.result['rewards']
no_context_result = task3.result
```

```python
plt.figure(3, figsize=(10,8))
plt.xlabel("Day")
plt.ylabel("Daily reward")

plt.plot(context_result['TS-Pricing'], label=f'C_TS-Pricing')
plt.plot(no_context_result['TS-Pricing'], label=f'NC_TS-Pricing')
plt.legend(loc='best')
plt.title("TS Comparison")
plt.show()
```

```python
plt.figure(4, figsize=(10,8))
plt.xlabel("Day")
plt.ylabel("\u0394 reward")

daily_gain = np.array(context_result['TS-Pricing']) - np.array(no_context_result['TS-Pricing'])
plt.plot(daily_gain)
plt.title("Context gain using a TS approach")
plt.show()
```
